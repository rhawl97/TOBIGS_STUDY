## 1. 'ì‚´ë¹ ì¡ŒJean' ë°”ì§€ êµ¬ë§¤ í›„ê¸° 1500ê°œ í¬ë¡¤ë§

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
base_url = 'https://a-bly.com/shop/goods/goods_view.php?goodsno=11209&page={}'
review_list = []

for n in range(20) :
    url = base_url.format(n+1)
    webpage = urlopen(url)
    source = BeautifulSoup(webpage, 'html5lib')
    reviews = source.select('article')
    
    for review in reviews :
        review_list.append(review.get_text().strip())
        #print(review.get_text().strip())
```

```python
review_list
```

```python
review_list = " ".join(review_list)
review_list
```

### (2) Tokenize

```python
from konlpy.tag import *
```

```python
twitter = Twitter()  #twitter class ê°ì²´ë‹¹!
komoran = Komoran()
kkma = Kkma()
```

```python
def make_corpus_rm_stopwords(text):
    corpus = []
    for s in text.split('\n'):
        corpus.append(['/'.join(p) for p in twitter.pos(s) if (p[1] != "Josa") & (p[1] != "Determiner") & (p[1] != "KoreanParticle") & (p[1] != "Punctuation")]) #ìž„ë² ë”©ì— í•„ìš”ì—†ëŠ” ì¡°ì‚¬ ë¹¼ê³ ! 
    return corpus
```

```python
corpus = make_corpus_rm_stopwords(review_list)
```

```python
corpus
```

## 3. Embedding

```python
from gensim.models import Word2Vec, fasttext

Skip_Gram_model = Word2Vec(corpus, size=25, window=5, min_count=5, workers=2, sg=1)
```

```python
def plot_with_labels(low_dim_embs, labels, filename='tsne.png'):
    assert low_dim_embs.shape[0] >= len(labels), 'More labels than embeddings'

    plt.figure(figsize=(18, 18))        # in inches

    # í•´ë‹¹ ì¢Œí‘œì— ì ì„ í‘œì‹œí•˜ê³ , ì˜¤ë¥¸ìª½/í•˜ë‹¨ ì •ë ¬ë¡œ ë‹¨ì–´ë¥¼ í‘œì‹œí•œë‹¤.
    for i, label in enumerate(labels):
        x, y = low_dim_embs[i]

        plt.scatter(x, y)
        plt.annotate(label,
                     xy=(x, y),
                     xytext=(5, 2),
                     textcoords='offset points',
                     ha='right',
                     va='bottom')

    plt.savefig(filename)
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# low_dim_embs ë‚´ë¶€ ê°¯ìˆ˜ì™€ n_componentsê°€ ê°™ì•„ì•¼ í•œë‹¤!
# n_components : ì°¨ì›
# perplexity : ê°€ìž¥ ê°€ê¹Œìš´ ì´ì›ƒ ê°¯ìˆ˜
# n_iter : ìµœì í™”ì— ì‚¬ìš©í•  ë°˜ë³µ íšŸìˆ˜
# init : embedding ì´ˆê¸°í™” ë°©ë²•
tsne = TSNE(perplexity=30, n_components=2, init='pca', n_iter=5000)

# plot ê°¯ìˆ˜. 8000ê°œì˜ embeddingsë¡œë¶€í„° ì•žìª½ ì¼ë¶€ë§Œ ì‚¬ìš©.
# low_dim_embsì—ëŠ” ë³€í™˜ëœ ì¢Œí‘œ x, yê°€ ë“¤ì–´ìžˆë‹¤.
plot_only = 500
low_dim_embs = tsne.fit_transform(vectors[:plot_only])     # (500, 2)
labels = words[:plot_only]                                  # ìž¬êµ¬ì„±í•œ ì½”ë“œ

plot_with_labels(low_dim_embs, labels)
```
